```python
import cv2
import numpy
```

```python
image = cv2.imread('median_blurred_image.jpg')
kenrel = numpy.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
```

```python
sharpen = cv2.filter2D(src=image, ddepth = -1, kernel = kenrel)
```

```python
cv2.imshow('Original Image', image)
cv2.imshow('Sharpened Image', sharpen)
cv2.waitKey(0)  
cv2.destroyAllWindows()
```

```python
import cv2
```

```python
image = cv2.imread('animate-social.webp', cv2.IMREAD_GRAYSCALE)
```

```python
canny_image = cv2.Canny(image=image, threshold1=50, threshold2=100)
```

```python
binary_threshold = cv2.threshold(src=image, thresh=200, maxval=255, type=cv2.THRESH_BINARY)[1]
```

```python
cv2.imshow('Original Image', image)
cv2.imshow('Canny Image', canny_image)
cv2.imshow('Binary Threshold Image', binary_threshold)
cv2.waitKey(0)  
cv2.destroyAllWindows()
```

```python
import cv2
```

```python
img = cv2.imread('hexagon.png')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
_, threshold = cv2.threshold(img, 200,255, cv2.THRESH_BINARY)
```

```python
contours, heirarchy = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for contour in contours:
    approx = cv2.approxPolyDP(contour, 0.01*cv2.arcLength(contour, True), True)

    corner = len(approx)

    if corner == 3:
        shape_name = "Triangle" 
    elif corner == 4:
        shape_name = "Quadrilateral"
    elif corner == 5:
        shape_name = "Pentagon"
    elif corner == 6:
        shape_name = "Hexagon"
    elif corner == 10:
        shape_name = "Star"
    else:
        shape_name = "Circle"
    cv2.drawContours(img, [approx], 0, (0,0,0), 5)
    cv2.putText(img, shape_name, (approx.ravel()[0], approx.ravel()[1]-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 2)
```

```python
cv2.imshow("Image 1", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    ret, frame = cap.read()

    if not ret:
        print('Could not read frame')
        break
    cv2.imshow('Webcam Feed', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        print('Quitting....')
        break
cap.release()
cv2.destroyAllWindows()
```

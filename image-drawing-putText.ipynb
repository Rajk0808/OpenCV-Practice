```python
import cv2
```

```python
image = cv2.imread('What-is-data-science-2.jpg')
```

```python
text_image = cv2.putText(img = image, text = 'Data Science', org = (50,50), fontFace = cv2.FONT_HERSHEY_SIMPLEX, fontScale = 1, color = (255,0,0), thickness = 2)
```

```python
cv2.imshow('Text Image', text_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

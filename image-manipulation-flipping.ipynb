```python
import cv2
```

```python
image = cv2.imread('What-is-data-science-2.jpg'
)
```

```python
flipped_image_vertically = cv2.flip(src = image, flipCode = 0)
cv2.imshow('Flipped Image Vertically', flipped_image_vertically)
```

```python
flipped_image_horizontally = cv2.flip(src = image, flipCode = 1)
cv2.imshow('Flipped Image Horizontally', flipped_image_horizontally)
```

```python
flipped_image_both = cv2.flip(src= image, flipCode = -1)
cv2.imshow('Flipped Image Both Vertically and Horizontally', flipped_image_both)
```

```python
cv2.waitKey(0)
cv2.destroyAllWindows()
```

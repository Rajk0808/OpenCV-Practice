```python
import cv2
import numpy as np
```

```python
img = np.zeros((500,500),dtype='uint8')
```

```python
rectangle = cv2.rectangle(img, (50,50), (250,250), (255), -1)
```

```python
_, img = cv2.threshold(rectangle, 200, 255, cv2.THRESH_BINARY)
```

```python
not_img = cv2.bitwise_not(img)
```

```python
contour, heirarchy = cv2.findContours(not_img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
not_img = cv2.drawContours(not_img, contour, -1,(255,0,0), 10)
```

```python
cv2.imshow("Image 1", not_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
```

```python
image = cv2.imread('What-is-data-science-2.jpg')
```

```python
center = (image.shape[1]//2, image.shape[0]//2)
```

```python
drawing_image = cv2.circle(img = image, center = center, radius = 100, color = (0,0,255), thickness = 5)
```

```python
cv2.imshow('Circle Drawned Image', drawing_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

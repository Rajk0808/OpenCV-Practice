```python
import cv2
```

```python
image = cv2.imread('What-is-data-science-2.jpg')
```

```python
drawing_image = cv2.line(img = image, pt1 =(0,0), pt2 = (250,250), color = (255,0,0), thickness = 5)
```

```python
cv2.imshow(winname='Line Drawned Image', mat=drawing_image)
```

```python
cv2.waitKey(0)
cv2.destroyAllWindows()
```
